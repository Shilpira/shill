```python
#we have to perform 2 sample 2 tail test
#here: H0: unit A = unit B
     # H1: unit A = unit B
```

```python
from scipy import stats
import pandas as pd
```

```python
BR = pd.read_csv("BuyerRatio.csv")
BR.head()
```

```python
x = BR.iloc[:,1:6]
```

```python
#as the given data was in categorical form and
#then converted into tabular format we will perform the chi_square test
#H0 = all proportion are euqal
#H1 = all proportion are not equal
```

```python
stats.chi2_contingency(x)[1]
```

```python
# p-value = 0.660
```

```python
#as we can see the p value is greater than 0.5
#so the H0 (null hypothesis is rejected)
#means all proportional are not equal
```
